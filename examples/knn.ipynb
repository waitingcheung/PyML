```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_moons
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

from lib.knn import KNeighborsClassifier
```

```python
X, y = make_moons(noise=0.3, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolor='black', s=40)
plt.show()
```

```python
# this classifier
model = KNeighborsClassifier(k=3)
y_pred = model.predict(X_train, y_train, X_test)
accuracy_score(y_test, y_pred)
```

```python
# sklearn classifier
from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train)
accuracy_score(y_test, neigh.predict(X_test))
```

```python
# plot the results    
np.random.seed(seed=0)

X_new, _ = make_moons(n_samples=2000, noise=0.5, random_state=0)
y_new = model.predict(X_train, y_train, X_new)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolor='black', s=40)
lim = plt.axis()
plt.scatter(X_new[:, 0], X_new[:, 1], c=y_new, cmap=plt.cm.Paired, s=20, alpha=0.1)

plt.axis(lim)
plt.show()
```
