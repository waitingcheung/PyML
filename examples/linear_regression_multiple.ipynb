```python
import random

import matplotlib.pyplot as plt
import numpy as np

from lib.linear_regression import LinearRegression
```

```python
rng = np.random.RandomState(1)
```

```python
X = rng.randn(200, 2)
y = np.dot(X, [-2, 1]) + 0.1 * rng.randn(X.shape[0])
```

```python
fig, ax = plt.subplots()
points = ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis')
ax.axis([-4, 4, -3, 3])
```

```python
# create a linear regression model
model = LinearRegression(verbose=True)
model.fit(X, y)
```

```python
model.weights
```

```python
# predict y from the data
X_new = rng.randn(100, 2)
y_new = model.predict(X_new)
```

```python
# plot the results
fig, ax = plt.subplots()
points = ax.scatter(X_new[:, 0], X_new[:, 1], c=y_new, s=50, cmap='viridis')
ax.axis([-4, 4, -3, 3])
```
